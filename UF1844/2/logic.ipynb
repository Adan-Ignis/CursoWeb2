1. Escriba un programa que solicite al usuario que ingrese una cadena.
El programa debería mostrar lo siguiente:

(a) El número total de caracteres de la cadena
(b) La cadena repetida 10 veces
(c) El primer carácter de la cadena (recuerde que los índices de cadena empiezan en 0)
(d) Los tres primeros caracteres de la cadena
(e) Los tres últimos caracteres de la cadena
(f) La cadena al revés
(g) El séptimo carácter de la cadena si la cadena es lo suficientemente larga; en caso contrario, un mensaje
(h) La cadena sin el primer y el último carácter
(i) La cadena en mayúsculas
(j) La cadena con cada a reemplazada por una e
(k) La cadena con cada letra reemplazada por un espacio

```python
import re

cadena = input("Escribe algo: ")

print(f"a) La cadena tiene {len(cadena)} caracteres")

for i in range(10):
    print(f"b) Interacción {i+1}: {cadena}")
# print((cadena + "\n") * 10) Hace lo mismo que el bucle

print(f"c) El primer carácter de la cadena es: {cadena[0]}")

print(f"d) Los 3 primeros caracteres de la cadena son: {cadena[:3]}")

print(f"e) Los 3 ultimos caracteres de la cadena son: {cadena[-3:]}")

print(f"f) La cadena al reves es: {cadena[::-1]}")

if len(cadena) > 6:
    print(f"g) El 7º carácter de la cadena es: {cadena[6]}")
else:
    print(f"g) La cadena tiene menos de 7 caracteres")

print(f"h) La cadena sin el primer y último carácter es: {cadena[1:-1]}")

print(f"i) La cadena en mayúscula es: {cadena.upper()}")

print(f'j) La cadena con cada "a" reemplazada por una "e" es: {cadena.replace("a", "e").cadena.replace("A", "E")} ')

kCadena = re.sub(r"[a-zA-Z]", "*", cadena)
print(f'k) La cadena con cada carácter alfabetico reemplazado por un "*" es: {kCadena}')
```

3. A menudo se olvida cerrar los paréntesis al introducir fórmulas.
Escriba un programa que solicite al usuario que introduzca una fórmula e imprima si esta tiene el mismo número de paréntesis de apertura y cierre.

```python
cadena = input("Escribe algo: ")

apertura = cadena.count("(")
cierre = cadena.count(")")

if apertura > cierre:
    print(f"El numero de paréntesis de apertura es mayor que de cierre")
elif apertura < cierre:
    print(f"El numero de paréntesis de apertura es menor que de cierre")
else:
    print(f"El numero de paréntesis de apertura y de cierre es el mismo")
```

7. Escriba un programa que solicite al usuario que ingrese una palabra y determine si es un palíndromo.
Un palíndromo es una palabra que se lee igual de derecha a izquierda.

```python
cadena = input("Escribe algo: ")

cadenaReves = cadena[::-1]

if cadena == cadenaReves:
    print(f"La palabra es un palíndromo: {cadena}")
else:
    print(f"La palabra no es un palíndromo")
    print(f"La palabra original es: {cadena}")
    print(f"La palabra del reves es: {cadenaReves}")
```

8. En una institución educativa, las direcciones de correo electrónico de los estudiantes terminan en @student.college.edu, mientras que las de los profesores terminan en @prof.college.edu.
Escriba un programa que primero pregunte al usuario cuántas direcciones de correo electrónico ingresará y luego le pida que las ingrese.
Una vez ingresadas todas las direcciones, el programa debería imprimir un mensaje indicando si todas las direcciones son de estudiantes o si se ingresaron algunas de profesores.

```python
ciclo = int(input("¿Cuantos direcciones de correo electronico ingresaras?: "))

correoElectronico = []
a = 0 # Contador de direcciones de estudiantes
b = 0 # Contador de direcciones de profesores

for i in range(ciclo):
    t = input("Dirección {i+1}º: ")
    correoElectronico.append(t)
    
for i in range(ciclo):
    if correoElectronico[i].endswitch("@student.college.edu"):
        a += 1
    elif correoElectronico[i].endswitch("@prof.college.edu"):
        b += 0

if a == 0 and b == 0:
    print("No hay direcciones de correo validas")
elif a >= 1 and b == 0:
    print("Todas las direcciones de correo son de estudiantes")
elif a >= 1 and b >= 1:
    print("Algunas direcciones de correo son de profesores")
elif a == 0 and b >= 1:
    print("Solamente hay direcciones de correo de profesores")
```
