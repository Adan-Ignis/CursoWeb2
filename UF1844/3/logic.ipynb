1. Escriba un programa que solicite al usuario ingresar una lista de números enteros. Haga lo siguiente:
(a) Imprima el número total de elementos de la lista.
(b) Imprima el último elemento de la lista.
(c) Imprima la lista en orden inverso.
(d) Imprima "Sí" si la lista contiene un 5 y "No" en caso contrario.
(e) Imprima el número de cincos en la lista.
(f) Elimine el primer y el último elemento de la lista, ordene los elementos restantes e imprima el resultado.
(g) Imprima cuántos números enteros de la lista son menores que 5.

```python
import random

lista_numero = []
valor_interaccion = 0

interaccion = int(random.uniform(1.0, 5.0))

for i in range(interaccion):
    valor_interaccion = input("Inserta el valor " + {i+1} + "º: ")
    lista_numero.append(valor_interaccion)

print("a)")
for i in lista_numero:
    print(f"Valor {i+1}: {lista_numero[i]}")

print(f"a) El número total de elementos es: len{lista_numero}")

print(f"b) El ultimo elemento es: {lista_numero[-1]}")

print(f"c) La lista en orden inverso es: {lista_numero[::-1]}")

if 5 in lista_numero:
    print("d) SI")
else:
    print("d) NO")

print(f"e) El numero de veces que se repite el 5 es de: lista_numero.count(5)")

nueva_lista = lista_numero[1:-1]
nueva_lista.sort()

print(f"f) La lista sin el primer y ultimo elemento es: {nueva_lista}")

contador = 0
for numero in lista_numero:
    if numero < 5:
        contador += 1

print(f"g) El numero de numeros menores a 5 es: {contador}")
```

2. Escriba un programa que genere una lista de 20 números aleatorios entre 1 y 100.
(a) Imprima la lista.
(b) Imprima el promedio de los elementos de la lista.
(c) Imprima el valor mayor y el menor de la lista.
(d) Imprima la segunda entrada mayor y la segunda menor de la lista.
(e) Imprima cuántos números pares hay en la lista.

```python
lista_numero = []

for i in range(20):
    lista_numero.append(random.uniform(1, 100))
    
# Hay un metodo para evitar que de problemas si la lista esta vacia
promedio = sum(lista_numero) / len(lista_numero)

numero_mayor = max(lista_numero)
numero_menor = min(lista_numero)

segundo_mayor = 0
segundo_menor = 0
numeros_pares = 0
for i in lista_numero:
    if i < numero_mayor and i > segundo_mayor:
        segundo_mayor = i
    if i > numero_menor and i < segundo_menor:
        segundo_menor = i
    if i % 2 == 0:
        numeros_pares += 1

print(f"La lista de numeros es: {lista_numero}")
print(f"El promedio de la lista es: {promedio}")
print(f"El valor mayor es: {numero_mayor}")
print(f"El valor menor es: {numero_menor}")
print(f"El segundo mayor valor es: {segundo_mayor}")
print(f"El segundo menor valor es: {segundo_menor}")
print(f"El numero de numeros pares de la lista es: {numeros_pares}")
```

4. Solicite al usuario que ingrese una lista con números entre 1 y 12.
Luego, reemplace todas las entradas de la lista mayores que 10 por 10.

```python
import random

Lista_numero = []
lista_10 = []

for i in range(24):
    lista_numero.append(random.uniform(1, 12))
    
for item in lista_numero:
    if lista_numero[item] > 10:
        lista_10.append(10)
    lista_10.append(lista_numero[item])

print(f"Lista original SI: {lista_numero}")
print(f"Lista original NO: {lista_10}")
```
